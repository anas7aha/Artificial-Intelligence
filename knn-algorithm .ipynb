```python
import warnings
warnings.filterwarnings('ignore')


#1- Import core libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
#  2- Download ready-made Iris data from sklearn
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['Species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
# 3- Display some rows of data
print(" Display the first 5 rows of data:")
print(df.head())
```

```python
#  4- Charts for data analysis
sns.pairplot(df, hue="Species")
plt.suptitle("ðŸ”¹Graph - Relationships between properties", y=1.02)
plt.show()

df.hist(figsize=(10, 6), edgecolor='black')
plt.suptitle("ðŸ”¹ Frequency tables for each property", y=1.02)
plt.tight_layout()
plt.show()
```

```python
#  5- Preparing data for KNN
X = df.drop("Species", axis=1)
y = df["Species"]
```

```python
# Encode text data if any (here the data is pre-encoded)
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
#  6- Data partitioning
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

```

```python
# 7- Create a KNN model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
# 8- Prediction and results
y_pred = knn.predict(X_test)
```

```python
print("\nâœ… Model accuracy:", accuracy_score(y_test, y_pred))
print("\nâœ… Classification Report:\n", classification_report(y_test, y_pred, target_names=iris.target_names))
```
